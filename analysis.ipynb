```python
import pandas as pd
```

```python
file_names = [
    "quarterly_Q615.csv",
    "quarterly_Q390.csv",
    "quarterly_Q161.csv",
    "quarterly_Q656.csv",
    "quarterly_Q602.csv",
    "quarterly_Q180.csv",
    "quarterly_Q667.csv",
    "quarterly_Q592.csv",
    "quarterly_Q144.csv",
    "quarterly_Q671.csv"
]
```

```python
data = pd.read_pickle('ALL_RESULT_m3_quarterly.pkl')
```

```python
for k in file_names:
    sample = data[k]
    mape_values = {key: value[2]for key, value in sample.items()}
    print(f"MAPE values for {k}: {mape_values}")
```

```python
new_dict = {}

for i, k in enumerate(file_names):
    if i==2 or i==3:
        continue    
    sample = data[k]
    for key, value in sample.items():
        if key not in new_dict:
            new_dict[key] = []
        new_dict[key].append(value[2])  # Append MAPE value
    
```

```python
mean_mape_values = {key: sum(values) / len(values) for key, values in new_dict.items()}
mean_mape_values
```
