```python
import pandas as pd
```

# Monthly

```python
mape_list = {}

models = ["deepar", "informer", "itransformer", "nbeats", "nhits", "patchtst"]

for model in models:
    data = pd.read_pickle(f"ALL_RESULT_{model}_m3_monthly.pkl")
    file_names = list(data["series_results"].keys())
    mape_list[model] = 0
    for file in file_names:
        mape_list[model] += data["series_results"][file]["metrics"]["mape"]
mape_list = {k: v / len(file_names) for k, v in mape_list.items()}
mape_list
```
